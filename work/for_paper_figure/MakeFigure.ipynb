```python
from math import pi
import random
import torch
from qiskit import QuantumCircuit
from qiskit.circuit.library import IGate, RZZGate

from gqco.task import define_tokens
from gqco.quantum import Circuit
```

```python
def gen_seqs(size, vocab_size, bad_tokens, seq_len=10, seed=373):
    random.seed(seed)
    
    seq = []
    while len(seq) < seq_len:
        val = random.randint(0, vocab_size)
        if val not in bad_tokens[size]:
            seq.append(val)
    return seq
```

```python
num_qubit = 20
size = 7
savefile='Circuit'
pool, bad_tokens, identity_id = define_tokens(num_qubit)
vocab_size = int(max(pool.keys(), key=int))+1
```

```python
tokens = gen_seqs(size=size, vocab_size=vocab_size, bad_tokens=bad_tokens, seq_len=20, seed=373)
```

## Circuit sampling

```python
tokens = gen_seqs(size=size, vocab_size=vocab_size, bad_tokens=bad_tokens, seq_len=20, seed=17)

myqc = Circuit('qiskit', num_qubit=size)
for t in tokens:
    for opr in pool[str(t)]:
        g = myqc.str_to_gate(opr['type'], theta=opr['theta'])
        myqc.qc.append(g, opr['control']+opr['loc'])
        

qc_figure = myqc.qc.draw(
    output='mpl', 
    style={'name': 'bw', 'backgroundcolor': '#FFFFFF'},  # Placeholder background (won't affect the final result)
    scale=0.35, 
    plot_barriers=False
)

display(qc_figure)

# Save the figure with a transparent background
qc_figure.figure.savefig(f'output/{savefile}-qc1.svg', transparent=True)
```

## VQE

```python
def make_qaoa_circuit(adj, layer_count):
    num_qubit = len(adj)
    
    qc = QuantumCircuit(num_qubit)

    for i in range(num_qubit):
        qc.h(i)

    for ell in range(layer_count):

        for i in range(num_qubit):
            w = adj[i][i].cpu().tolist()
            if w != 0:
                qc.rz(w, i)

        for i in range(num_qubit-1):
            for j in range(i+1, num_qubit):
                w = adj[i, j].cpu().tolist()
                if w != 0:
                    qc.rzz(w, i, j)
        qc.barrier()
            
        for k in range(num_qubit):
            qc.rx(2, k)
    
    return qc
```
